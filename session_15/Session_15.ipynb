* 1. Install scikit-learn if you haven't already, then load the iris dataset and fit a GradientBoostingClassifier to classify the flower species. Print the model's accuracy on the test set.

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
df
```

### Train, Test, Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.2, random_state=42)
```

### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)

print(f"Accuracy Score: {acc:.4f}")
```

* 2. Using the Flipkart product review dataset (or any small CSV with positive/negative labels), train an AdaBoostClassifier to predict if a review is positive or negative. Show the classification report (precision, recall, f1-score) for your model.<br><br><em><strong>Hint:</strong> You can use sklearn's AdaBoostClassifier and train_test_split.</em>

```python
review = [
    "Amazing product, super fast delivery and great battery backup!",
    "Worst experience ever. The phone stopped working in two days.",
    "Value for money, highly recommended product.",
    "Very poor quality. Disappointed with Flipkart packaging.",
    "The camera quality is brilliant, love this product.",
    "Horrible service, defective item delivered.",
    "Decent product for the price. Happy with it.",
    "Waste of money. Do not buy this product."
    ]
sentiment = [1, 0, 1, 0, 1, 0, 1, 0]  # 1 = Positive, 0 = Negative
```

```python
# Text Preprocessing (Converting text data to numeric vectors)
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_rev  = tfidf.fit_transform(review)
```

```python
from sklearn.model_selection import train_test_split

X_train_rev, X_test_rev, y_train_rev, y_test_rev = train_test_split(X_rev, sentiment, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(n_estimators=50, random_state=42)
```

```python
ada_clf.fit(X_train_rev, y_train_rev)
```

```python
ada_preds = ada_clf.predict(X_test_rev)
```

```python
from sklearn.metrics import classification_report
```

```python
print("AdaBoost Classification Report:\n")
print(classification_report(y_test_rev, ada_preds, target_names=['Negative', 'Positive']))
```

* 3. Explain in your own words the difference between a weak learner and a strong learner, and give one real-world example of each from apps you use (like Instagram, Zomato, or Paytm).

#### Weak vs Strong Learners

* Weak Learner: A simple model that performs just slightly better than random guessing. It only looks at one tiny piece of information and makes a quick, very basic rule.
* Strong Learner: A complex, highly accurate model created by combining hundreds of weak learners together to look at the big picture.

#### Real-World Example (Instagram Feeds):

* Weak Learner: A simple rule that says: "If the user scrolled past this video in 1 second, they didn't like it." This is too simple and often wrong (maybe they sneezed!).
* Strong Learner: The actual Instagram algorithm, which combines the scroll speed rule + comment sentiment + watch time + audio saves, forming a highly accurate prediction of what you want to see next.

* 4. Modify your GradientBoostingClassifier code to tune the n_estimators and learning_rate parameters. Try at least 3 different combinations and report which one gives the best accuracy.<br><br><em><strong>Hint:</strong> n_estimators controls the number of boosting stages, learning_rate shrinks the contribution of each tree.</em>

```python
n_estimators_list = [10, 50, 100]
learning_rate_list = [0.01, 0.1, 0.2]
```

```python
best_acc = 0
best_params = None
```

```python
for n in n_estimators_list:
    for lr in learning_rate_list:
        model = GradientBoostingClassifier(n_estimators=n, learning_rate=lr, random_state=42)
        model.fit(X_train, y_train)
        
        acc = accuracy_score(y_test, model.predict(X_test))
        print(f"Trees: {n:3} | LR: {lr:4} | Accuracy: {acc:.4f}")
        
        if acc > best_acc:
            best_acc = acc
            best_params = (n, lr)
```

```python
print("Best Combination")
print(f"Trees: {best_params[0]}, Learning Rate: {best_params[1]}, Accuracy: {best_acc:.4f}")
```

#### Boosting in Paytm (Fraud Detection)

* How it improves performance:
Fraudsters constantly change their tactics to trick basic security systems. A Gradient Boosting model works perfectly here because it trains sequentially—each new "tree" specifically focuses on correcting the mistakes (false positives or false negatives) made by the previous trees. This creates an incredibly adaptive and highly accurate security net that catches complex, evolving fraudulent transactions in real-time while letting normal payments pass instantly.
