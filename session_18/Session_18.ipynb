* 1. Load the Iris dataset using scikit-learn and use cross_val_score() with a KNeighborsClassifier to compute the average accuracy using 5-fold cross-validation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
knn_scores = cross_val_score(knn_model, X_iris, y_iris, cv=5)
print(f"Average 5-fold Accuracy: {knn_scores.mean():.4f}")
```

* 3. Use StratifiedKFold from scikit-learn to split the Iris dataset and compare the distribution of target classes in each fold to show how stratified sampling works.<br><br><em><strong>Hint:</strong> Print the class counts for each fold to verify stratification.</em>

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
```

```python
for fold_idx, (train_idx, test_idx) in enumerate(skf.split(X_iris, y_iris), 1):
    y_test_fold = y_iris[test_idx]
    unique, counts = np.unique(y_test_fold, return_counts=True)
    print(f"Fold {fold_idx} Class Distribution: {dict(zip(unique, counts))}")
```

* 4. Experiment with different values of 'cv' (3, 5, 10) in cross_val_score() for a RandomForestClassifier on the Wine dataset and report which value gives the most stable (least variable) accuracy scores.<br><br><em><strong>Hint:</strong> Look at the standard deviation of the scores for each cv value.</em>

```python
from sklearn.datasets import load_wine

wine = load_wine()
```

```python
X_wine = wine.data
y_wine = wine.target
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
```

```python
cv_options = [3, 5, 10]

for cv_val in cv_options:
    scores = cross_val_score(rf_model, X_wine, y_wine, cv=cv_val)
    print(f"CV= {cv_val:2} | Mean: {scores.mean():.4f} | Std Dev: {scores.std():.4f}")
```
