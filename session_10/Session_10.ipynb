* 1. Create a Python function sigmoid(x) that takes a numeric input and returns the sigmoid value using the formula 1 / (1 + exp(-x)). Test your function with x = -2, 0, and 3.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))   # Calculates the sigmoid of a numeric input or numpy array.
```

```python
test_values = [-2, 0, 3]
print("Task 1: Sigmoid Function Tests:\n")
for x in test_values:
    print(f"sigmoid({x}) = {sigmoid(x):.4f}")
```

* 2. Given a dataset of 10 Instagram posts with features 'likes' and 'has_caption' (1 if caption present, 0 if not), and a label 'viral' (1 if post went viral, 0 otherwise), write code to fit a logistic regression model to predict 'viral' using scikit-learn. Print the model coefficients.

```python
data = {
    "likes": [150, 2000, 50, 3500, 400, 5000, 100, 1200, 300, 4500],
    "has_caption": [1, 1, 0, 1, 0, 1, 0, 1, 1, 0],
    "viral": [0, 1, 0, 1, 0, 1, 0, 0, 0, 1],
}

df = pd.DataFrame(data)
```

```python
X = df[["likes", "has_caption"]]
y = df["viral"]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X, y)
```

```python
print("Task 2: Logistic Regression Coefficients")
print(f"Intercept: {model.intercept_[0]:.4f}")
print(f"Coefficient for 'likes': {model.coef_[0][0]:.6f}")
print(f"Coefficient for 'has_caption': {model.coef_[0][1]:.4f}")
```

```python
print(model.coef_)
```

* 3. Plot the sigmoid function for x values from -10 to 10 using matplotlib, and visually indicate the threshold at 0.5 on the plot.<br><br><em><strong>Hint:</strong> Use plt.axhline() to draw the threshold line.</em>

```python
# Generate x values from -10 to 10
x_vals = np.linspace(-10, 10, 200)
y_vals = sigmoid(x_vals)
```

```python
# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="Sigmoid Function", color="blue", linewidth=2)

# Visually indicate the threshold at 0.5
plt.axhline(y=0.5, color="red", linestyle="--", label="Threshold (0.5)",)
plt.axvline(x=0, color="gray", linestyle=":", alpha=0.7)  # Reference line where sigmoid(x) = 0.5

# Aesthetics
plt.title("Sigmoid Function with Decision Threshold")
plt.xlabel("x")
plt.ylabel("Probability / Output")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

* 4. Given a probability output from your logistic regression model, write a function that classifies the result as 'Viral' or 'Not Viral' based on a threshold value that you can set (default 0.5). Test your function with probabilities 0.2, 0.6, and 0.8 and thresholds 0.5 and 0.7.

```python
def classify_probability(prob, threshold=0.5):
    if prob >= threshold:       # Classifies a probability output as 'Viral' or 'Not Viral' based on a threshold.
        return "Viral"
    else:
        return "Not Viral"
```

```python
# Testing parameters
probabilities = [0.2, 0.6, 0.8]
thresholds = [0.5, 0.7]
```

```python
print("Task 4: Classification Testing")
for t in thresholds:
    print(f"\nUsing Threshold: {t}")
    # print("-" * 25)                                 # for printing dash line
    for p in probabilities:
        result = classify_probability(p, threshold=t)
        print(f"Probability: {p} = Class: {result}")
```
