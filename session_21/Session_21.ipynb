* 1. Load a CSV file containing daily temperature data for Ahmedabad and create two new features: a 3-day lag of the temperature and a 7-day rolling average of the temperature using pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ahm = pd.read_csv(r"C:\Users\patel prit\OneDrive\Desktop\Assessment\ML\Assignment\Supervised\Session - 21\ahmedabad_daily_temperature.csv")

df_ahm['temperature'] = pd.to_numeric(df_ahm['temperature'], errors='coerce')

df_ahm['date'] = pd.to_datetime(df_ahm['date'], format='%Y-%m-%d')
df_ahm.set_index('date', inplace=True)
df_ahm.sort_index(inplace=True)

df_ahm = df_ahm.dropna()
```

```python
df_ahm['Lag_3'] = df_ahm['temperature'].shift(3)
df_ahm['Rolling_7'] = df_ahm['temperature'].rolling(window=7).mean()

print(df_ahm.head(10))
```

* 3. Train a Random Forest Regressor to forecast the next day's temperature using the lag and rolling features you created earlier. Print the predicted value for the next available date.

```python
df_clean = df_ahm.dropna()

X = df_clean[['Lag_3', 'Rolling_7']]
y = df_clean['temperature']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=50, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
latest_data = df_clean.iloc[-1][['Lag_3', 'Rolling_7']].values.reshape(1, -1)
next_day_pred = rf_model.predict(latest_data)

next_date = df_clean.index[-1] + pd.Timedelta(days=1)
print(f"Predicted Max Temperature for {next_date.strftime('%Y-%m-%d')}: {next_day_pred[0]:.2f}°C")
```

* 4. Calculate MAE, RMSE, and MAPE between your model's predictions and the actual values for the test set. Print all three metrics and explain in one line which metric is most sensitive to large errors.<br><br><em><strong>Hint:</strong> Use sklearn.metrics.mean_absolute_error, mean_squared_error, and write a simple function for MAPE.</em>

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

y_preds = rf_model.predict(X_test)

mae = mean_absolute_error(y_test, y_preds)
rmse = np.sqrt(mean_squared_error(y_test, y_preds))
mape = calculate_mape(y_test, y_preds)

print(f"MAE:  {mae:.4f}°C")
print(f"RMSE: {rmse:.4f}°C")
print(f"MAPE: {mape:.4f}%")
```

* Explanation: RMSE (Root Mean Squared Error) is the most sensitive to large errors because it squares every error before averaging them, heavily punishing any massive outliers!

* 5. Use ChatGPT or Copilot to help you write code for calculating SMAPE (Symmetric Mean Absolute Percentage Error) for your time series predictions. Paste the code you used and mention one thing you learned from the AI's explanation.

```python
def calculate_smape(y_true, y_pred):
    """
    Calculates the Symmetric Mean Absolute Percentage Error.
    """
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return 100 * np.mean(diff)
```

```python
smape_val = calculate_smape(y_test.values, y_preds)
print(f"SMAPE: {smape_val:.4f}%")
```

##### What I learned: 

* I learned that standard MAPE heavily penalizes the model when actual values are very close to zero, creating huge artificial errors. SMAPE completely fixes this by taking the absolute average of both the true and predicted values in the denominator, making it totally symmetric and much more stable for datasets with values nearing zero.
