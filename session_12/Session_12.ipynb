* 1. Load the Iris dataset using scikit-learn, train a DecisionTreeClassifier, and plot the resulting decision tree using plot_tree().

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
df
```

# Split: X, y

```python
X = pd.DataFrame(df.data, columns=df.feature_names)
y = df.target
```

### Train, Test, Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### 

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='gini', random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 10))
plot_tree(model, feature_names=df.feature_names, class_names=df.target_names, filled=True, rounded=True)
plt.title("Decision Tree using Gini Criterion (Full Depth)")
plt.show()
```

2. Change the splitting criterion of your DecisionTreeClassifier from 'gini' to 'entropy' and compare the accuracy scores for both on the same dataset.<br><br><em><strong>Hint:</strong> Use the 'criterion' parameter when initializing the classifier.</em>

```python
model2 = DecisionTreeClassifier(criterion='entropy', random_state=42)
model2.fit(X_train, y_train)
```

```python
y_pred_gini = model.predict(X_test)
y_pred_entropy = model2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_gini = accuracy_score(y_test, y_pred_gini)
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
```

```python
print(f"Accuracy using 'Gini' criterion: {accuracy_gini}")
print(f"Accuracy using 'Entropy' criterion: {accuracy_entropy}")
```

3. Limit the maximum depth of your decision tree to 2 and visualize the new tree structure. Observe and briefly describe how the tree's decisions have changed compared to the full-depth tree.

```python
model3= DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=42)
model3.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(10, 6))
plot_tree(model3, feature_names=df.feature_names, class_names=df.target_names, filled=True, rounded=True)
plt.title("Decision Tree Limited to Max Depth = 2")
plt.show()
```

## Observation:
The full-depth tree created many complex splits to perfectly classify every single training sample (which can lead to overfitting). By limiting max_depth=2, the tree is forced to stop early. It only makes two levels of decisions. While it might misclassify a few edge cases (leaves are no longer completely "pure"), it creates a much simpler, more generalized rule set that is easier to interpret.
