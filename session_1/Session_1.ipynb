### 1. Netflix recommending movies based on your watch history
Type: Supervised Learning

Simple Explanation:
Netflix looks at the movies and shows you watched before and learns what kind of content you like. Then it predicts which movies you may like next.

### 2. Spotify grouping similar songs into playlists
Type: Unsupervised Learning

Simple Explanation:
Spotify finds songs that are similar in music style, mood, or genre and puts them together in playlists. Nobody tells the system which songs belong together; it finds the patterns by itself.


### 3. A self-driving car learning to park by trial and error
Type: Reinforcement Learning

Simple Explanation:
The car tries different parking actions. When it parks correctly, it gets a reward. When it makes mistakes, it gets a penalty. Over time, it learns the best way to park.

Hint: Focus on rules vs learning from data.

## Task 4: Traditional Programming vs Machine Learning for a WhatsApp Spam Filter

A traditional programming approach would require developers to manually create rules to identify spam messages. For example, if a message contains specific keywords such as "Win Money", "Lottery", or "Free Prize", the program would mark it as spam. However, maintaining these rules can be difficult because spammers continuously change their messages.

In contrast, a machine learning approach learns from a large dataset of spam and non-spam messages. Instead of manually defining rules, the model identifies patterns in the data and automatically learns how to distinguish spam from legitimate messages. As new examples are provided, the model can improve its performance and adapt to new spam techniques more effectively than traditional programming.
