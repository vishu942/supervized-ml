* 4. Take a Zomato restaurant dataset and use scikit-learn's RFE (Recursive Feature Elimination) with a DecisionTreeRegressor to select the 3 most important features for predicting 'average cost for two'. List the selected features and the RFE ranking.

```python
df = pd.read_csv(r"C:\Users\patel prit\OneDrive\Desktop\Assessment\ML\Assignment\Supervised\Session - 4\zomato_restaurants.csv")
```

```python
df.head()
```

```python
X = df.drop('Average_Cost_For_Two', axis=1)
y = df['Average_Cost_For_Two']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE

model = DecisionTreeRegressor(random_state=42)

rfe = RFE(estimator=model, n_features_to_select=3)      # RFE = find the most important features (feature selection technique)

rfe.fit(X,y)
```

```python
selected_features = X.columns[rfe.support_]
print("Selected Features:\n", selected_features)
```

```python
print("RFE Ranking:")

for feature, rank in zip(X.columns, rfe.ranking_):
    print(feature, ":", rank)
```

* 5. Given a BookMyShow movie ratings dataset, detect outliers in the 'user ratings' column using the IQR method. Remove the outliers and display how many rows were dropped.

Hint: Calculate Q1, Q3, and IQR; filter out rows outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR].

```python
df = pd.read_csv(r"C:\Users\patel prit\OneDrive\Desktop\Assessment\ML\Assignment\Supervised\Session - 4\bookmyshow_ratings.csv")
```

```python
df.head()
```

```python
col = 'User_Ratings'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df_clean = df[
    (df[col] >= lower_bound) &
    (df[col] <= upper_bound)
]

rows_dropped = len(df) - len(df_clean)

print("\nOriginal Rows:", len(df))
print("Rows After Removing Outliers:", len(df_clean))
print("Rows Dropped:", rows_dropped)

print(df_clean.head())
```
