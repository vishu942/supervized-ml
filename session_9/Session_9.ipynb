* 1. Download a publicly available dataset of your choice (for example, a Flipkart product ratings dataset or Spotify song features) and split it into features (X) and target (y) for a regression problem. 

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\patel prit\OneDrive\Desktop\Assessment\ML\Assignment\Supervised\Session - 2\Spotify.csv", encoding="latin1")
```

```python
df.head()
```

```python
X = df.select_dtypes(include=['number']).drop(columns=['Column1', 'Popularity'], errors='ignore')
y = df['Popularity']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
feature_names = X.columns

print("Features selected for the model:\n", list(feature_names))
```

* 2. Fit a Lasso regression model (L1 regularization) using scikit-learn's Lasso class on your dataset, and print out the coefficients for each feature.

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha=1.0, random_state=42)
lasso_model.fit(X_scaled, y)
```

```python
print("Lasso Regression Coefficients:")
lasso_coeffs = pd.DataFrame({
    'Feature': feature_names,
    'Lasso Coefficient': lasso_model.coef_
})
lasso_coeffs
```

* 3. Fit a Ridge regression model (L2 regularization) using scikit-learn's Ridge class on the same dataset, and compare the coefficients to those from the Lasso model.<br><br><em><strong>Hint:</strong> Look for features where Lasso sets coefficients exactly to zero but Ridge does not.</em>

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=5.0, random_state=42)
ridge_model.fit(X_scaled, y)
```

```python
print("--- Comparison: Lasso vs. Ridge ---")
comparison_df = pd.DataFrame({
    'Feature': feature_names,
    'Lasso (L1)': lasso_model.coef_,
    'Ridge (L2)': ridge_model.coef_
})
print(comparison_df.to_string(index=False))
```

### Observation:"
Notice how Lasso sets features like Beats.Per.Minute, Energy and Danceability exactly to 0.0,
while Ridge shrinks them but keeps all of them non-zero.

* 4. Fit an ElasticNet regression model using scikit-learn's ElasticNet class, and observe how the coefficients differ from both Lasso and Ridge.<br><br><em><strong>Hint:</strong> Try changing the l1_ratio parameter and see how it affects the results.</em>

```python
from sklearn.linear_model import ElasticNet

# Fit ElasticNet with a balanced mix (l1_ratio = 0.5)
enet_balanced = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)
enet_balanced.fit(X_scaled, y)
```

```python
# Fit ElasticNet leaning more towards Ridge (l1_ratio = 0.2)
enet_ridge_heavy = ElasticNet(alpha=1.0, l1_ratio=0.2, random_state=42)
enet_ridge_heavy.fit(X_scaled, y)
```

```python
print("ElasticNet Performance with different l1_ratios:")
enet_df = pd.DataFrame({
    'Feature': feature_names,
    'Lasso (alpha=1.0)': lasso_model.coef_,
    'ElasticNet (l1=0.5)': enet_balanced.coef_,
    'ElasticNet (l1=0.2)': enet_ridge_heavy.coef_,
    'Ridge (alpha=5.0)': ridge_model.coef_
})
enet_df
```

* 5. Suppose you want to build a feature selector for a Zomato-style restaurant rating predictor. Use Lasso regression to select the most important features from your dataset, and list which features are kept (non-zero coefficients) and which are dropped.

```python
# Filter out features based on whether their Lasso coefficient is 0 or not
kept_features = feature_names[lasso_model.coef_ != 0].tolist()
dropped_features = feature_names[lasso_model.coef_ == 0].tolist()

print("Feature Selection Results using Lasso")
print(f"\n✅ Kept Features (Non-zero coefficients): {len(kept_features)}")
for col in kept_features:
    coef = lasso_model.coef_[feature_names.get_loc(col)]
    print(f"  - {col} (Weight: {coef:.4f})")

print(f"\n❌ Dropped Features (Coefficients forced to 0): {len(dropped_features)}")
for col in dropped_features:
    print(f"  - {col}")
```
